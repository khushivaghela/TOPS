```python
#1) What is Power BI and how does it differ from Excel?
"""Power BI is a business intelligence application used to visualize data and build interactive reports and dashboards. It's perfect for working with big datasets,
reporting automation, and linking numerous data sources.Excel, on the other hand, is more focused on spreadsheet calculations and data analysis but does not have the advanced
visualization and automation that Power BI has. Although Excel is excellent for dealing with structured data and formulas,
Power BI is more suitable for taking and analyzing big, complex datasets with real-time refresh."""
```

.

```python
#2) Explain the concept of data modeling in Power BI.
""" Power BI data modeling is all about structuring your data in the most efficient manner by establishing the relationships between tables, defining calculated columns and
measures, and restructuring data for analytics. It serves to make the data more valuable and easier to handle while designing reports and dashboards."""
```

.

```python
#3) What are the different types of connections available in Power BI?
"""Power BI primarily provides two types of connections:
- Import Mode – Imports data into Power BI, increasing report speeds but necessitating refreshes from time to time.
- DirectQuery Mode – Does not store data in the source and queries it in real-time, providing real-time updates but potentially slower.
There is also a Live Connection to connect to some databases such as SSAS (SQL Server Analysis Services) & a Composite Model to have a combination of Import and DirectQuery."""
```

.

```python
#4) How do you handle data transformation in Power BI?
"""Data transformation in Power BI is done using Power Query, where you can clean and shape data before loading it into the report.
This includes removing duplicates, changing data types, merging queries, and adding calculated columns."""
```

.

```python
#5) What is DAX and why is it important?
"""DAX (Data Analysis Expressions) is a formula language in Power BI used to perform calculations and create custom metrics.
It helps in aggregating data, calculating percentages, running totals, and implementing logic-based functions for deeper insights."""
```

.

```python
#6)Calculated columns vs. Measures?
"""Calculated columns are created at the row level and stored in the data model. They are useful when a new field needs to be added to a table.
Measures are calculated dynamically based on user interactions and filters. They do not consume extra storage and are more efficient for large datasets."""
```

.

```python
#7)How to handle table relationships?
"""Table relationships in Power BI are managed in the Model View. Relationships can be one-to-many, one-to-one, or many-to-many.
You can define cross-filtering directions and use functions like RELATED or LOOKUPVALUE to fetch data from connected tables"""
```

.

```python
#8) Purpose of Power BI Gateway?
"""A Power BI Gateway acts as a bridge between on-premises data sources (such as SQL Server or Excel files) and the Power BI Service.
It allows users to securely refresh data without manually uploading it to the cloud."""
```

.

```python
#9) How to schedule data refresh?
"""Data refresh can be scheduled in Power BI Service through the dataset settings.
Users can set automatic refresh intervals (daily, weekly, etc.) and use a Power BI Gateway for refreshing data from on-premises sources."""
```

.

```python
#10) What is Row-Level Security (RLS)?
"""Row-Level Security (RLS) restricts data access based on user roles. It ensures that users only see the data relevant to them.
RLS is implemented using DAX filters in Power BI Desktop and assigned to users in Power BI Service."""
```

.

```python
#11) Power BI Desktop vs. Power BI Service?
"""Power BI Desktop is used for building and designing reports, including data modeling and visualization.
Power BI Service is the cloud platform where reports are published, shared, and refreshed.
The Service enables collaboration, scheduling refreshes, and setting up security controls."""
```

.

```python
#12) Explain the concept of Direct Query in Power BI.
"""DirectQuery allows Power BI to directly connect to a data source without importing the data.
This means reports always show real-time or near-real-time data.
It's useful for large datasets where importing everything would be inefficient.
However, it can be slower than Import Mode since every interaction runs a query on the database."""
```

.

```python
#13) What are Power BI templates and how are they useful?
"""Power BI templates (.PBIT) allow you to save a report structure without including the data.
They store visuals, queries, and model relationships so you can reuse them with different datasets.
They help standardize reporting and save time when working on multiple reports with similar formats."""
```

.

```python
#14) How do you handle incremental data refresh in Power BI?
"""Incremental refresh loads only new or changed data instead of refreshing the entire dataset.
This improves performance, especially for large datasets.
It is set up in Power BI Desktop using parameters for range-based partitioning and then published to Power BI Service."""
```

.

```python
#15) What is the role of Power Query in Power BI?
"""Power Query is used for data transformation before loading it into Power BI.
It helps clean, reshape, merge, and filter data from different sources.
It also automates data preparation, making reports more efficient and accurate."""
```

.

```python
#16) Explain the difference between calculated columns and calculated tables in Power BI.
"""A calculated column is added to an existing table and works at the row level.
A calculated table is a new table created from existing data using DAX, often used for summary views or relationships.
Columns are useful for additional fields, while tables help with structural changes in the model."""
```

.

```python
#17) How do you create custom visuals in Power BI?
"""Custom visuals can be created using Power BI’s SDK with TypeScript and D3.js.
Alternatively, Power BI Marketplace offers pre-built custom visuals.
Developers build custom visuals for specific needs when built-in options are not sufficient."""
```

.

```python
#18) What are the best practices for optimizing performance in Power BI?
"""Use Import Mode instead of DirectQuery for better speed when possible.
Reduce the number of visuals and limit data points in charts.
Optimize DAX calculations and remove unnecessary columns.
Use aggregations, indexing, and data compression to enhance performance."""
```

.

```python
#19) Power BI integrates with Azure services like Azure SQL, Synapse Analytics, and Data Lake for cloud storage and analytics.
"""It connects with Office 365 tools like Excel, SharePoint, and Teams for collaboration.
Power Automate and Power Apps can also be used for automation and interactive data applications."""
```

.

```python
#20) Explain the concept of aggregations in Power BI.
"""Aggregations pre-calculate and store summarized data to improve performance.
Instead of querying large tables, Power BI retrieves smaller pre-aggregated data, making reports faster.
It is useful for large datasets where detailed queries would be slow."""
```

.

```python
#21) How do you handle error handling and data quality in Power BI?
"""Use Power Query to clean and validate data before loading it into reports.
Check for missing values, incorrect data types, and duplicates.
DAX functions like IFERROR and ISBLANK can be used to handle errors in calculations."""
```

.

```python
#22) What is the purpose of Power BI Embedded and when would you use it?
"""Power BI Embedded allows developers to integrate Power BI reports into custom applications.
It is used when businesses want to offer interactive analytics within their own software or web applications.
This allows users to view reports without needing a Power BI account."""
```
