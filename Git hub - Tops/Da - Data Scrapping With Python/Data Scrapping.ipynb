```python
import numpy as np
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
#single page

urll = "https://www.mywishcare.com/?utm_source=googleads&utm_medium=googleads&utm_campaign=brand_searchads&gad_source=1&gclid=CjwKCAiA2JG9BhAuEiwAH_zf3tCFF1RhY9DqhEAMnWmrXVsrVk-MGgXFCpQslydQSn_IeJpTES7cjBoCkaoQAvD_BwE"
response = requests.get(urll)
response
```

```python
data = response.text
soup = BeautifulSoup(data, 'html.parser')
soup
```

```python
soup.prettify()
```

```python
for x in soup.find_all("a", class_ = "product-title h6 line-clamp"):
  print(x.text.strip())
```

```python
for y in soup.find_all("p", class_ = "line-clamp"):
  print(y.text.strip())
```

```python
for z in soup.find_all(class_ = "h6 text-subdued line-through"):
  print(z.text.strip())
```

```python
product = []
for x in soup.find_all("a", class_ = "product-title h6 line-clamp"):
  product.append(x.text.strip())


product
```

```python
work = []
for x in soup.find_all("p", class_ = "line-clamp"):
  work.append(x.text.strip())


work
```

```python
price = []
for x in soup.find_all(class_ = "h6 text-subdued line-through"):
  price.append(x.text.strip())


price = price[0::1]
price
```

```python
data = {'product': product, 'work': work, 'price': price}
df = pd.DataFrame(data)
df
```

```python
# @title price

from matplotlib import pyplot as plt
import seaborn as sns
df.groupby('price').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
##multiple pages

import requests
from bs4 import BeautifulSoup

# Remove the angle brackets from the base_url
urll = "https://www.mywishcare.com/?utm_source=googleads&utm_medium=googleads&utm_campaign=brand_searchads&gad_source=1&gclid=CjwKCAiA2JG9BhAuEiwAH_zf3tCFF1RhY9DqhEAMnWmrXVsrVk-MGgXFCpQslydQSn_IeJpTES7cjBoCkaoQAvD_BwE"
num_pages_to_scrape = 6

for page_num in range(1, num_pages_to_scrape + 1):

  url = urll.format(page_num)

  response = requests.get(urll)

  if response.status_code == 200:

    soup = BeautifulSoup(response.text, 'html.parser')

    articles = soup.find_all("a",class_="product-title h6 line-clamp")

    for article in articles:
      print(article.text.strip())

    else:
      print(f"Failed to retrieve page {page_num}")
```
