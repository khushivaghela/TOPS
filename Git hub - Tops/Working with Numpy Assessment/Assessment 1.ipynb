```python
#6) What are the top 5 states with the most failed banks?
banks.groupby('ST')['Bank Name'].count().sort_values(ascending=False).head(5)
```

```python
#7) What are the top 5 acquiring institutions?
banks.groupby('Acquiring Institution')['Bank Name'].count().sort_values(ascending=False).head(5)
```

```python
#8) How many banks has the State Bank of Texas acquired? How many of them were actually in Texas?
# Total acquisitions by State Bank of Texas
total_acquisitions = banks[banks['Acquiring Institution'] == 'State Bank of Texas'].shape[0]

# Acquisitions in Texas by State Bank of Texas
texas_acquisitions = banks[(banks['Acquiring Institution'] == 'State Bank of Texas') & (banks['ST'] == 'TX')].shape[0]

print(f"Total acquisitions by State Bank of Texas: {total_acquisitions}")
print(f"Acquisitions in Texas by State Bank of Texas: {texas_acquisitions}")
```

```python
#9) What is the most common city in California for a bank to fail in?
california_banks = banks[banks['ST'] == 'CA']
most_common_city = california_banks.groupby('City')['Bank Name'].count().idxmax()
print(f"The most common city in California for a bank to fail in is: {most_common_city}")
```
